```python
archivo = open("caida.xyz", "w")    # posiciones (formato 3D)
archiva = open("caida.dat", "w")    # datos numéricos (tiempo, altura, velocidad)

# Pedimos los valores al usuario
x0 = float(input("x0 (posición inicial horizontal): "))   # normalmente 0
y0 = float(input("y0 (altura inicial en metros): "))       # por ejemplo 10
n = int(input("npasos (cuántos pasos de tiempo): "))       # por ejemplo 500

# Constantes físicas
sigmar = 3.16   # escala (puede ser un factor de conversión)
grave = 9.8     # gravedad (m/s²)
dt = 0.005      # intervalo de tiempo (s)

# Condiciones iniciales
tiempo = 0.0
velocidad = 0.0

# Escribimos encabezado en caida.dat
archiva.write("tiempo(s)\taltura(m)\tvelocidad(m/s)\n")

# Bucle principal: simulamos paso a paso
for i in range(n + 1):

    # cálculo de la posición y velocidad
    y = y0 - 0.5 * grave * (tiempo ** 2)
    velocidad = -grave * tiempo     # v = -g * t (caída libre)
    y_real = y * sigmar             # escala (opcional)

    # Guardamos en caida.xyz
    archivo.write("1\n")
    archivo.write(" \n")
    archivo.write(f"{x0} {y_real}\n")
    archivo.write(f"H {x0} {y_real} 0.0\n")

    # Guardamos en caida.dat
    archiva.write(f"{tiempo:.3f}\t{y:.3f}\t{velocidad:.3f}\n")

    # Avanzamos el tiempo
    tiempo += dt

# Cerramos los archivos
archivo.close()
archiva.close()

# Mensaje final
print("\n Simulación completada")
print("Archivo 'caida.xyz' generado (coordenadas 3D)")
print("Archivo 'caida.dat' generado (tiempo, altura, velocidad)")
```
