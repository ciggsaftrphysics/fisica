```python
## Saisie des coefficients
a = float(input("Entrez a (différent de 0) : "))
b = float(input("Entrez b : "))
c = float(input("Entrez c : "))

# Calcul du discriminant
delta = b**2 - 4*a*c

if delta > 0:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print(f"Deux solutions réelles : x1 = {x1}, x2 = {x2}")
    
elif delta == 0:
    x = -b / (2*a)
    print(f"Une solution réelle double : x = {x}")
else:
    # Solutions complexes
    real = -b / (2*a)
    imag = math.sqrt(-delta) / (2*a)
    print(f"Deux solutions complexes : {real} + {imag}i et {real} - {imag}i")
```
